```python
# combine dataframes as needed 

from functools import reduce
import pandas as pd

# list of baldal market area dataframes 
bd_pop_list = [
    bd_pop_2017, bd_pop_2018, bd_pop_2019, bd_pop_2020,
    bd_pop_2021, bd_pop_2022, bd_pop_2023, bd_pop_2024
]

# merge all dataframes on the same keys
bd_pop = reduce(lambda left, right: pd.merge(left, right, on=["행정구역", "시군구코드"], how="inner"), bd_pop_list)
# add a column that indicates the market area type
bd_pop["상권유형"] = "발달상권"
# get rid of decimal marks in the population columns 
bd_pop = bd_pop.replace({',': ''}, regex=True)
# filter for "고양시(goyang city)"
bd_pop = bd_pop[bd_pop["행정구역"] == "고양시"]

# list of golmok market area dataframes
gm_pop_list = [
    gm_pop_2017, gm_pop_2018, gm_pop_2019, gm_pop_2020,
    gm_pop_2021, gm_pop_2022, gm_pop_2023, gm_pop_2024
]   
# merge all dataframes on the same keys
gm_pop = reduce(lambda left, right: pd.merge(left, right, on=["행정구역", "시군구코드"], how="inner"), gm_pop_list)
# add a column that indicates the market area type
gm_pop["상권유형"] = "골목상권"
# get rid of decimal marks in the population columns 
gm_pop = gm_pop.replace({',': ''}, regex=True)
# filter for "고양시(goyang city)"
gm_pop = gm_pop[gm_pop["행정구역"] == "고양시"]

# list of jeontong market area dataframes
jt_pop_list = [
    jt_pop_2017, jt_pop_2018, jt_pop_2019, jt_pop_2020,
    jt_pop_2021, jt_pop_2022, jt_pop_2023, jt_pop_2024
]
# merge all dataframes on the same keys
jt_pop = reduce(lambda left, right: pd.merge(left, right, on=["행정구역", "시군구코드"], how="inner"), jt_pop_list) 
# add a column that indicates the market area type
jt_pop["상권유형"] = "전통시장"
# get rid of decimal marks in the population columns 
jt_pop = jt_pop.replace({',': ''}, regex=True)
# filter for "고양시(goyang city)"
jt_pop = jt_pop[jt_pop["행정구역"] == "고양시"]

# concatenate all market area dataframes into a single dataframe
combined_pop = pd.concat([bd_pop, gm_pop, jt_pop], ignore_index=True)
combined_pop
```

```python
# filter 길단위 columns only 
streetpop_columns = [col for col in combined_pop.columns if "길단위" in col]
filtered_streetpop_data = combined_pop[["행정구역", "시군구코드", "상권유형"] + streetpop_columns]

# transverse the columns and rows for better readability
filtered_streetpop_data = filtered_streetpop_data.melt(id_vars=["행정구역", "시군구코드", "상권유형"], var_name="연도_분기", value_name="길단위 인구")   

# make a year columns and quarter columns using the '연도_분기' column
filtered_streetpop_data[['연도', '분기']] = filtered_streetpop_data['연도_분기'].str.extract(r'(\d{4})년(\d+분기)')
del filtered_streetpop_data['연도_분기']
# convert '길단위 인구' to numeric type
filtered_streetpop_data['길단위 인구'] = pd.to_numeric(filtered_streetpop_data['길단위 인구'], errors='coerce')
# sum up the population by year, quarter, and market area type 
filtered_streetpop_data = filtered_streetpop_data.groupby(['연도', '상권유형']).agg({'길단위 인구': 'sum'}).reset_index()
# drop values for year 2025 since it only contains data for 2분기
filtered_streetpop_data = filtered_streetpop_data[filtered_streetpop_data['연도'] != '2025']
filtered_streetpop_data 
```

```python
# plot the opening counts by year and market area type
plt.figure(figsize=(12, 6))
sns.lineplot(data=filtered_streetpop_data, x='연도', y='길단위 인구', hue='상권유형', marker='o')
plt.title('고양시 연도별 상권유형별 길단위 인구 추이', fontsize=16)
plt.xlabel('년도')
plt.ylabel('길단위 인구')
plt.legend(title='상권유형')
plt.show()
```
