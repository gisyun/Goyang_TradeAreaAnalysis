```python
sales = pd.read_excel('./기업_매출액.xlsx')
employ = pd.read_excel('./기업_종사자수.xlsx')

print(sales.head())
print(employ.head())
```

```python
# replace rows with '*' and '-' to NaN
sales = sales.replace(['*', '-'], np.nan)
employ = employ.replace(['*', '-'], np.nan)
# convert data type to float
sales = sales.astype({'전체기업_소계_2022':'float', '중소기업_소계_2022':'float', '소기업_소계_2022':'float',
                      '소기업_소상공인_2022':'float', '소기업_소상공인제외_2022':'float', '중기업_소계_2022':'float',
                      '중소기업 외_소계_2022':'float', '중소기업_소계_2023':'float', '소기업_소계_2023':'float',
                      '소기업_소상공인_2023':'float', '소기업_소상공인제외_2023':'float', '중기업_소계_2023':'float'})

employ = employ.astype({'중소기업_소계_2022':'float', '소기업_소계_2022':'float',
                      '소기업_소상공인_2022':'float', '소기업_소상공인제외_2022':'float', '중기업_소계_2022':'float',
                      '중소기업외_2022':'float', '중소기업_소계_2023':'float', '소기업_소계_2023':'float',
                      '소기업_소상공인_2023':'float', '소기업_소상공인제외_2023':'float', '중기업_소계_2023':'float'})
```

```python
sales.info()
```

```python
employ.info()
```

```python
# filter columns for 소기업_소상공인제외 and 중기업_소계 for 2022 and 2023 

sales = sales[['시군구명', '산업별(10차)대분류별', '소기업_소상공인제외_2022', '중기업_소계_2022',
               '소기업_소상공인제외_2023', '중기업_소계_2023']]

employ = employ[['시군구명', '산업별(10차)대분류별', '소기업_소상공인제외_2022', '중기업_소계_2022',
               '소기업_소상공인제외_2023', '중기업_소계_2023']]

sales
```

```python
# groupyby 산업별(10차)대분류별 and sum the values for each year
sales_grouped = sales.groupby('산업별(10차)대분류별').sum().reset_index()
employ_grouped = employ.groupby('산업별(10차)대분류별').sum().reset_index()

print(sales_grouped.head())
print(employ_grouped.head())
```

```python
sales_grouped
```

```python
# drop 시군구명 column 
sales_grouped = sales_grouped.drop(columns=['시군구명'])
employ_grouped = employ_grouped.drop(columns=['시군구명'])

sales_grouped.head()
```

```python
sales_grouped.to_excel('./기업_매출액_산업대분류별.xlsx', index=False)
employ_grouped.to_excel('./기업_고용_산업대분류별.xlsx', index=False)
```
