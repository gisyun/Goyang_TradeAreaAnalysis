```python
# combine dataframes as needed 

from functools import reduce
import pandas as pd

# list of baldal market area dataframes 
bd_nbsr_list = [
    bd_nbsr_2017, bd_nbsr_2018, bd_nbsr_2019, bd_nbsr_2020,
    bd_nbsr_2021, bd_nbsr_2022, bd_nbsr_2023, bd_nbsr_2024, bd_nbsr_2025
]

# merge all dataframes on the same keys
bd_nbsr = reduce(lambda left, right: pd.merge(left, right, on=["행정구역", "시군구코드"], how="inner"), bd_nbsr_list)
# add a column that indicates the market area type
bd_nbsr["상권유형"] = "발달상권"
# filter for "고양시(goyang city)"
bd_nbsr = bd_nbsr[bd_nbsr["행정구역"] == "고양시"]

# list of golmok market area dataframes
gm_nbsr_list = [
    gm_nbsr_2017, gm_nbsr_2018, gm_nbsr_2019, gm_nbsr_2020,
    gm_nbsr_2021, gm_nbsr_2022, gm_nbsr_2023, gm_nbsr_2024, gm_nbsr_2025
]   
# merge all dataframes on the same keys
gm_nbsr = reduce(lambda left, right: pd.merge(left, right, on=["행정구역", "시군구코드"], how="inner"), gm_nbsr_list)
# add a column that indicates the market area type
gm_nbsr["상권유형"] = "골목상권"
# filter for "고양시(goyang city)"
gm_nbsr = gm_nbsr[gm_nbsr["행정구역"] == "고양시"]

# list of jeontong market area dataframes
jt_nbsr_list = [
    jt_nbsr_2017, jt_nbsr_2018, jt_nbsr_2019, jt_nbsr_2020,
    jt_nbsr_2021, jt_nbsr_2022, jt_nbsr_2023, jt_nbsr_2024, jt_nbsr_2025
]
# merge all dataframes on the same keys
jt_nbsr = reduce(lambda left, right: pd.merge(left, right, on=["행정구역", "시군구코드"], how="inner"), jt_nbsr_list) 
# add a column that indicates the market area type
jt_nbsr["상권유형"] = "전통시장"
# filter for "고양시(goyang city)"
jt_nbsr = jt_nbsr[jt_nbsr["행정구역"] == "고양시"]

# concatenate all market area dataframes into a single dataframe
combined_nbsr = pd.concat([bd_nbsr, gm_nbsr, jt_nbsr], ignore_index=True)
combined_nbsr
```

```python
# filter 1년 생존율 columns only 
survival_columns = [col for col in combined_nbsr.columns if "1년" in col]
filtered_survival_data = combined_nbsr[["행정구역", "시군구코드", "상권유형"] + survival_columns]

# transverse the columns and rows for better readability
filtered_survival_data = filtered_survival_data.melt(id_vars=["행정구역", "시군구코드", "상권유형"], var_name="연도_분기", value_name="생존율(1년)")   

# make a year column and quarter column using the '연도_분기' column
filtered_survival_data[['연도', '분기']] = filtered_survival_data['연도_분기'].str.extract(r'(\d{4})년(\d+분기)')
del filtered_survival_data['연도_분기']

# average 생존율 for each year and market area type 
filtered_survival_data = filtered_survival_data.groupby(['연도', '상권유형']).agg({'생존율(1년)': 'mean'}).reset_index()

# extract the values of year 2025
filtered_survival_data = filtered_survival_data[filtered_survival_data['연도'] != '2025']
filtered_survival_data 
```

```python
# plot the survival rates for each market area type over the years
plt.figure(figsize=(10, 6))
sns.lineplot(data=filtered_survival_data, x='연도', y='생존율(1년)', hue='상권유형', marker='o')
plt.title('고양시 연도별 신생기업 생존율(1년) 추이 (%)')
plt.xlabel('연도')
plt.ylabel('생존율 (%)')
plt.legend(title='상권유형')
plt.show()
```

```python
# combine dataframes as needed 

from functools import reduce
import pandas as pd

# list of service dataframes 
service_nbsr_list = [
    service_nbsr_2017, service_nbsr_2018, service_nbsr_2019, service_nbsr_2020,
    service_nbsr_2021, service_nbsr_2022, service_nbsr_2023, service_nbsr_2024
]

# merge all dataframes on the same keys
service_nbsr = reduce(lambda left, right: pd.merge(left, right, on=["행정구역", "지역코드"], how="inner"), service_nbsr_list)
# add a column that indicates the market area type
service_nbsr["업종"] = "서비스업"
# filter for "고양시(goyang city)"
service_nbsr = service_nbsr[service_nbsr["행정구역"] == "고양시"]

# list of retail dataframes
retail_nbsr_list = [
    retail_nbsr_2017, retail_nbsr_2018, retail_nbsr_2019, retail_nbsr_2020,
    retail_nbsr_2021, retail_nbsr_2022, retail_nbsr_2023, retail_nbsr_2024
]
# merge all dataframes on the same keys
retail_nbsr = reduce(lambda left, right: pd.merge(left, right, on=["행정구역", "지역코드"], how="inner"), retail_nbsr_list)
# add a column that indicates the market area type
retail_nbsr["업종"] = "소매업"
# filter for "고양시(goyang city)"
retail_nbsr = retail_nbsr[retail_nbsr["행정구역"] == "고양시"]

# list of food dataframes
food_nbsr_list = [
    food_nbsr_2017, food_nbsr_2018, food_nbsr_2019, food_nbsr_2020,
    food_nbsr_2021, food_nbsr_2022, food_nbsr_2023, food_nbsr_2024
]
# merge all dataframes on the same keys
food_nbsr = reduce(lambda left, right: pd.merge(left, right, on=["행정구역", "지역코드"], how="inner"), food_nbsr_list)
# add a column that indicates the market area type
food_nbsr["업종"] = "음식업"
# filter for "고양시(goyang city)"
food_nbsr = food_nbsr[food_nbsr["행정구역"] == "고양시"]

# concatenate all market area dataframes into a single dataframe
combined_nbsr = pd.concat([service_nbsr, retail_nbsr, food_nbsr], ignore_index=True)
combined_nbsr
```

```python
# filter 1년 생존율 columns only 
survival_columns = [col for col in combined_nbsr.columns if "1년" in col]
filtered_survival_data = combined_nbsr[["행정구역", "지역코드", "업종"] + survival_columns]

# transverse the columns and rows for better readability
filtered_survival_data = filtered_survival_data.melt(id_vars=["행정구역", "지역코드", "업종"], var_name="연도_분기", value_name="생존율(1년)")   

# make a year column and quarter column using the '연도_분기' column
filtered_survival_data[['연도', '분기']] = filtered_survival_data['연도_분기'].str.extract(r'(\d{4})년(\d+분기)')
del filtered_survival_data['연도_분기']

# average 생존율 for each year and market area type 
filtered_survival_data = filtered_survival_data.groupby(['연도', '업종']).agg({'생존율(1년)': 'mean'}).reset_index()

# extract the values of year 2025
filtered_survival_data = filtered_survival_data[filtered_survival_data['연도'] != '2025']
filtered_survival_data 
```
