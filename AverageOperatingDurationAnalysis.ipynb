```python
# combine dataframes as needed 

from functools import reduce
import pandas as pd

# list of baldal market area dataframes 
bd_opdur_list = [
    bd_opdur_2017, bd_opdur_2018, bd_opdur_2019, bd_opdur_2020,
    bd_opdur_2021, bd_opdur_2022, bd_opdur_2023, bd_opdur_2024
]

# merge all dataframes on the same keys
bd_opdur = reduce(lambda left, right: pd.merge(left, right, on=["행정구역", "시군구코드"], how="inner"), bd_opdur_list)
# add a column that indicates the market area type
bd_opdur["상권유형"] = "발달상권"
# get rid of decimal marks in the population columns
bd_opdur = bd_opdur.replace({',': ''}, regex=True)
# filter for "고양시(goyang city)"
bd_opdur = bd_opdur[bd_opdur["행정구역"] == "고양시"]

# list of golmok market area dataframes
gm_opdur_list = [
    gm_opdur_2017, gm_opdur_2018, gm_opdur_2019, gm_opdur_2020,
    gm_opdur_2021, gm_opdur_2022, gm_opdur_2023, gm_opdur_2024
]   
# merge all dataframes on the same keys
gm_opdur = reduce(lambda left, right: pd.merge(left, right, on=["행정구역", "시군구코드"], how="inner"), gm_opdur_list)
# add a column that indicates the market area type
gm_opdur["상권유형"] = "골목상권"
# get rid of decimal marks in the population columns
gm_opdur = gm_opdur.replace({',': ''}, regex=True)
# filter for "고양시(goyang city)"
gm_opdur = gm_opdur[gm_opdur["행정구역"] == "고양시"]

# list of jeontong market area dataframes
jt_opdur_list = [
    jt_opdur_2017, jt_opdur_2018, jt_opdur_2019, jt_opdur_2020,
    jt_opdur_2021, jt_opdur_2022, jt_opdur_2023, jt_opdur_2024
]
# merge all dataframes on the same keys
jt_opdur = reduce(lambda left, right: pd.merge(left, right, on=["행정구역", "시군구코드"], how="inner"), jt_opdur_list)
# add a column that indicates the market area type
jt_opdur["상권유형"] = "전통시장"
# get rid of decimal marks in the population columns 
jt_opdur = jt_opdur.replace({',': ''}, regex=True)
# filter for "고양시(goyang city)"
jt_opdur = jt_opdur[jt_opdur["행정구역"] == "고양시"]

# concatenate all market area dataframes into a single dataframe
combined_opdur = pd.concat([bd_opdur, gm_opdur, jt_opdur], ignore_index=True)
combined_opdur
```

```python
# filter 10년 평균 영업기간 columns only
tenyear_columns = [col for col in combined_opdur.columns if "10년" in col]
filtered_opdur_data = combined_opdur[["행정구역", "시군구코드", "상권유형"] + tenyear_columns]

# transverse the columns and rows for better readability
filtered_opdur_data = filtered_opdur_data.melt(id_vars=["행정구역", "시군구코드", "상권유형"], var_name="연도_분기", value_name="평균영업기간(10년)")   

# make a year columns and quarter columns using the '연도_분기' column
filtered_opdur_data[['연도', '분기']] = filtered_opdur_data['연도_분기'].str.extract(r'(\d{4})년(\d+분기)')
del filtered_opdur_data['연도_분기']
# sum up opening counts by year and market area type
filtered_opdur_data = filtered_opdur_data.groupby(['연도', '상권유형']).agg({'평균영업기간(10년)': 'sum'}).reset_index()
# drop values for year 2025 since it only contains data for 2분기
filtered_opdur_data = filtered_opdur_data[filtered_opdur_data['연도'] != '2025']
filtered_opdur_data
```

```python
# plot the average operation duration(ten years) by year and market area type
plt.figure(figsize=(12, 6))
sns.lineplot(data=filtered_opdur_data, x='연도', y='평균영업기간(10년)', hue='상권유형', marker='o')
plt.title('고양시 연도별 상권유형별 평균영업기간(10년) 추이', fontsize=16)
plt.xlabel('년도')
plt.ylabel('평균영업기간(10년)')
plt.legend(title='상권유형')
plt.show()
```

```python
# filter 30년 평균 영업기간 columns only
thirtyyear_columns = [col for col in combined_opdur.columns if "30년" in col]
filtered_opdur_data = combined_opdur[["행정구역", "시군구코드", "상권유형"] + thirtyyear_columns]

# transverse the columns and rows for better readability
filtered_opdur_data = filtered_opdur_data.melt(id_vars=["행정구역", "시군구코드", "상권유형"], var_name="연도_분기", value_name="평균영업기간(30년)")   

# make a year columns and quarter columns using the '연도_분기' column
filtered_opdur_data[['연도', '분기']] = filtered_opdur_data['연도_분기'].str.extract(r'(\d{4})년(\d+분기)')
del filtered_opdur_data['연도_분기']
# sum up opening counts by year and market area type
filtered_opdur_data = filtered_opdur_data.groupby(['연도', '상권유형']).agg({'평균영업기간(30년)': 'sum'}).reset_index()
# drop values for year 2025 since it only contains data for 2분기
filtered_opdur_data = filtered_opdur_data[filtered_opdur_data['연도'] != '2025']
filtered_opdur_data
```
